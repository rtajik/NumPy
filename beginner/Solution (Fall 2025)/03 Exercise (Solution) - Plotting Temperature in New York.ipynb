# 03 Exercise - Plotting Temperature in New York

### from Fahrenheit (F) to Celsius (C)
$$ C = 5 / 9 (F - 32) $$

```python
# Don't modify this cell, only run it.

# Import NumPy
import numpy as np

# making the dataset of temperature in Celsius
record_low_list = [-28, -22, -21, 9, 25, 35, 40, 34, 24, 16, -11, -22]
record_low_array = np.array(record_low_list)
record_low_celsius = 5 / 9 * (record_low_array - 32)

# Finding the min, max and mean
min_temp = np.min(record_low_celsius)
max_temp = np.max(record_low_celsius)
argmax_temp = np.argmax(record_low_celsius)
mean_temp = np.mean(record_low_celsius)

# Printing out the information.
print("The temperature in Celsius are: \n", record_low_celsius.round(2), "\n")
print("The minimum temperature is: \n", min_temp.round(2), "\n")
print("The maximum temperature is: \n", max_temp.round(2), "\n")
print("The index of maximum temperature is: \n", argmax_temp.round(2), "\n")
print("The average temperature is: \n", mean_temp.round(2), "\n")
```

```python
# The array of month
year_month = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December",
]
year_month = np.array(year_month)
```

```python
# Import matplotliob
import matplotlib.pyplot as plt
```

```python
# Make the bar plot
plt.figure(figsize=(15, 5))
plt.bar(year_month, record_low_celsius)

# Write the labels and title
plt.xlabel("Months")
plt.ylabel("Degrees")
plt.title("Record low temperature on year in Celsius")

# Show the plot
plt.show()
```

## Exercise 4: Approximating Temperature Using Cosine in 3 steps


### Exercise 4a 
#### The amplitude is defined to be
$$ amplitude = (max - min)/2 $$

```python
# Computing the amplitude
amp = (max_temp - min_temp) / 2
print(amp.round(2))

# Months with highest temperature
# month_highest_temp = year_month[argmax_temp]
# print(month_highest_temp)
# # Or
# month_highest_temp = year_month[record_low_celsius.argmax()]
# print(month_highest_temp)
# Or
month_highest_temp = year_month[record_low_celsius == record_low_celsius.max()][0]
print(month_highest_temp)
```

### Exercise 4b 
#### Find the vector
$$ 2 \pi (months_{temp} - max)/12 $$

```python
# Computing the vector describe above
months_num = np.arange(12)
v = 2 * np.pi / 12 * (months_num - argmax_temp)
```

### Exercise 4c 
#### Find an approximating for temperature each month by using the formula
$$ amplitude * cos(2 \pi ((months_{number} - 1) - max_{index})/12) + mean_{temp} $$

```python
# Computing the approximated temperatures
approx = amp * np.cos(v) + mean_temp
print(approx.round(2))
```

## Exercise 5: Plot the Difference

```python
# Take the absolute value of difference
Cosine = np.cos(v)

# Plot the difference in a bar plot
plt.figure(figsize=(15, 5))
plt.plot(year_month, record_low_celsius, label="low in celsius")
plt.plot(year_month, approx, label="approximated in celsius")
plt.xlabel("Months")
plt.ylabel("Degrees")
plt.title("Record low temperature on year in Celsius")
plt.legend()
plt.show()
```

```python
# Take the absolute value of difference
error = np.abs(record_low_celsius - approx)
print(error)

# Plot the difference in a bar plot
plt.figure(figsize=(15, 5))
plt.bar(year_month, error, label="error in celsius")
plt.xlabel("Months")
plt.ylabel("Degrees")
plt.title("Record low temperature on year in Celsius")
plt.legend()
plt.show()
```

## Exercise 6: Plot the Function
### Plot the function
$$ f(x) = amplitude * cos(2 \pi (x - max_{index})/12) + mean_{temp} $$

### Assume we have a 30-day period. Write a program that, under the assumption that temperature varies continuously over time, can approximate (estimate) the temperature for each day.

```python
# Make daily forecasting
num_approx = 12*30
daily_num = np.arange(num_approx)
v = 2 * np.pi / num_approx * (daily_num - (argmax_temp+1)*30)
months_day = 30*np.ones((1, 12))
days_passed = months_day.cumsum()

# Compute the y_value
daily_approx = amp * np.cos(v) + mean_temp

# Plot the function
plt.figure(figsize=(15, 5))

plt.bar(daily_num + 1, daily_approx, label="$daily_{Approximated}$ in celsius")
plt.bar(
    days_passed,
    approx,
    label="$month_{Approximated}$ in celsius",
)

plt.xlabel("Months")
plt.ylabel("Degrees")
plt.title("Record low temperature on year in Celsius")
plt.legend()

# Show the function
plt.show()
```
